```python
import cv2 as cv
import numpy as np
import sys
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

img=cv.imread(file_path + '/image_set/lena_color.png')

if img is None:
  sys.exit('File Not Found')

"""
OpenCV -> BGR image
"""

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

"""
cv.GaussianBlur(src, dst=None, ksize, sigmaX, sigmaY=None, bordertype=None)
- src    : source
- dst    : destination
- ksize  : kernel size.
- sigmaX : x-direction sigma
- sigmaY : y-direction sigma
- borderType
"""
# src, ksize
dst = cv.GaussianBlur(gray,(5,5),1)
dst2 = cv.GaussianBlur(gray,(5,5),3)
dst3 = cv.GaussianBlur(gray,(0,0),7)


cv2_imshow(gray)
cv2_imshow(dst)
cv2_imshow(dst2)
cv2_imshow(dst3)
```
