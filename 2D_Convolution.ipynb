```python
import cv2 as cv
import numpy as np
import sys
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

img=cv.imread(file_path + '/image_set/lena_color.png')
gray=cv.imread(file_path + '/image_set/lena_gray.png')

if img is None:
  sys.exit('File Not Found')

"""
OpenCV -> BGR image
"""

"""
cv2.filter2D(src, dst=None, ddepth, kernel, anchor=None, delta=None, borderType=None)
- src        : source
- dst        : destination
- ddepth     : dest data type (-1을 지정하면 같은 타입의 dst 생성)
- kernel     : filter mask matrix
- anchor     : 고정점 위치. (-1, -1)이면 필터 중앙 고정점
- delta      : 추가적으로 더할 값
- borderType : 가장자리 픽셀 확장 방식
"""

kernel = np.ones((5,5), np.float32)/25
dst = cv.filter2D(img, -1, kernel)
dst_gray = cv.filter2D(gray, -1, kernel)


cv2_imshow(img)
cv2_imshow(dst)
cv2_imshow(gray)
cv2_imshow(dst_gray)
```

```python
"""
matplotlib -> RGB image
"""

RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.float32)/25
dst2 = cv.filter2D(RGB_img, -1, kernel)
dst2_gray = cv.filter2D(gray, -1, kernel)


plt.subplot(121), plt.imshow(RGB_img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(dst2), plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()

plt.subplot(121), plt.imshow(gray), plt.title('Original Gray')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(dst2_gray), plt.title('Averaging Gray')
plt.xticks([]), plt.yticks([])
plt.show()

cv.waitKey()
cv.destroyAllWindows()
```
